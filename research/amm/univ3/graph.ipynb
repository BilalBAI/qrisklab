```python
import os
from dotenv import load_dotenv
load_dotenv()
api_key = os.getenv('GRAPH_API_KEY')

from qrisklab.clients.graph import GraphClient
from qrisklab.clients.graph_query import (
    UNISWAP_V3_ENDPOINT,
    UNISWAP_V4_ENDPOINT,
    v3_pools_query,
    v4_overview_query,
)
```

```python
# Uniswap V3 — top pools
client = GraphClient(endpoint=UNISWAP_V3_ENDPOINT, api_key=api_key)
client.query(query=v3_pools_query(first=10))
```

```python
# Uniswap V4 — overview
v4_client = GraphClient(endpoint=UNISWAP_V4_ENDPOINT, api_key=api_key)
v4_client.query(query=v4_overview_query())
```

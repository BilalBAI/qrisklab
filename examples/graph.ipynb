```python
import os
from dotenv import load_dotenv
load_dotenv()
API_KEY = os.getenv('GRAPH_API_KEY')

from qrisklab.clients.graph import GraphClient, UNISWAP_V3_ENDPOINT, UNISWAP_V3_QUERY_EXAMPLE, UNISWAP_V4_ENDPOINT, UNISWAP_V4_QUERY_EXAMPLE    

api_key = API_KEY
```

```python
# Uniswap V3
client = GraphClient(endpoint=UNISWAP_V3_ENDPOINT, api_key=api_key)
client.query(query=UNISWAP_V3_QUERY_EXAMPLE)
```

```python
# Uniswap V4
client = GraphClient(endpoint=UNISWAP_V4_ENDPOINT, api_key=api_key)
client.query(query=UNISWAP_V4_QUERY_EXAMPLE)
```
